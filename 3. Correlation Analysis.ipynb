### Spearman Correlation

```python
columns_of_interest = ['hearing loss rate','POI density','population','sex ratio']
spearman_corr_matrix = data1[columns_of_interest].corr(method='spearman')

sns.set(style="white")
plt.figure(figsize=(4.5, 4))
ax = sns.heatmap(spearman_corr_matrix, annot=True, cmap='coolwarm', linewidths=.5, fmt=".2f",annot_kws={'size': 10})
#plt.title('Spearman Correlation Matrix')
plt.xticks(rotation=90)
plt.yticks(rotation=0)  
ax.tick_params(axis='x', labelsize=10)  
ax.tick_params(axis='y', labelsize=10) 

plt.tight_layout()
```
