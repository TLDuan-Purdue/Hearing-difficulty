```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib_inline.backend_inline

# 设置Retina显示
matplotlib_inline.backend_inline.set_matplotlib_formats('retina')

# 设置字体为 Helvetica
plt.rcParams['font.family'] = 'Helvetica'

columns_of_interest = ['Hearing loss rate','Overall POI density','Risky POI density','road density','population','Index of relative ruruality','median income', 'poverty fraction','uninsured fraction','sex ratio','old dependency ratio','% over bachelor','% below highschool']

# 计算 Spearman 相关矩阵
spearman_corr_matrix = data1[columns_of_interest].corr(method='spearman')

# 创建一个上三角形的遮罩
mask = np.triu(np.ones_like(spearman_corr_matrix, dtype=bool))

sns.set(style="white")

# 创建图形
plt.figure(figsize=(14, 11))

ax = sns.heatmap(spearman_corr_matrix, 
                 mask=mask, 
                 annot=True, 
                 cmap='coolwarm', 
                 linewidths=.5, 
                 fmt=".2f",
                 annot_kws={'size': 15},
                 cbar_kws={"shrink": .5})

# 设置刻度标签旋转角度和字体大小
plt.xticks(rotation=90, ha='right')
plt.yticks(rotation=0)  
ax.tick_params(axis='x', labelsize=15)  
ax.tick_params(axis='y', labelsize=15) 

# 调整布局以适应图像内容
plt.tight_layout()

# 保存高分辨率图像
plt.savefig('/Users/duan114/Library/CloudStorage/OneDrive-purdue.edu/24 hl/paper writing/Images/AAA/f3.png', dpi=300, bbox_inches='tight', format='png')

# 显示图像
plt.show()
```
